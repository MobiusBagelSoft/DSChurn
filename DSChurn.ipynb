```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
def object_to_int(dataframe_series):
    if dataframe_series.dtype=='object':
        dataframe_series = LabelEncoder().fit_transform(dataframe_series)
    return dataframe_series
```

```python


df = df.apply(lambda x: object_to_int(x))
df.head()

```

```python
df = df.drop(['customerID'], axis = 1)
```

```python
df.head()
```

```python


X = df.drop(columns = ['Churn'])
y = df['Churn'].values

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.30, random_state = 1, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier(n_estimators=1000 , n_jobs = -1,)
model_rf.fit(X_train, y_train)

rfprediction_test = model_rf.predict(X_test)
```

```python
from sklearn.metrics import recall_score, confusion_matrix, precision_score, f1_score, accuracy_score, classification_report
```

```python
print(classification_report(y_test, rfprediction_test))
```

```python
print(accuracy_score(y_test, rfprediction_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors = 11) 
knn_model.fit(X_train,y_train)
predicted_y = knn_model.predict(X_test)
accuracy_knn = knn_model.score(X_test, y_test)
```

```python
print(classification_report(y_test, predicted_y))
print(accuracy_score(y_test, predicted_y))
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression(max_iter=2000)
lr_model.fit(X_train , y_train)
```

```python
lr_pred= lr_model.predict(X_test)
print(classification_report(y_test, lr_pred))
print(accuracy_score(y_test, lr_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtmodel = DecisionTreeClassifier()
dtmodel.fit(X_train,y_train)
predictdt_y = dtmodel.predict(X_test)
accuracy_dt = dtmodel.score(X_test, y_test)
```

```python
print(classification_report(y_test, predictdt_y))
print(accuracy_score(y_test, predictdt_y))
```

```python
import tensorflow as tf
nn = tf.keras.models.Sequential()
nn.add(tf.keras.layers.Dense(units=19, activation='relu'))

nn.add(tf.keras.layers.Dense(units=19, activation='relu'))
# nn.add(tf.keras.layers.Dropout(0.1))
nn.add(tf.keras.layers.Dense(units=15, activation='relu'))

nn.add(tf.keras.layers.Dense(units=10, activation='relu'))
# nn.add(tf.keras.layers.Dropout(0.1))
nn.add(tf.keras.layers.Dense(units=8, activation='relu'))

# nn.add(tf.keras.layers.Dense(units=1, activation='tanh')) #Output Layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

nn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
history=nn.fit(X_train, y_train, batch_size = 16, epochs = 100, validation_split = 0.1)
```

```python
from tensorflow.keras.utils import plot_model
plot_model(nn, show_shapes = True)
```

```python
nnypred=nn.predict(X_test)
```

```python
nnypred2 = [1 if x > 0.5 else 0 for x in nnypred ]
```

```python
print(classification_report(y_test, nnypred2))
print(accuracy_score(y_test, nnypred2))
```

```python
def model_performance_graphs2():
    
    fig, axes = plt.subplots(1, 2, figsize = (15, 8))

    axes[0].plot(history.epoch, history.history['accuracy'], label = 'acc')
    axes[0].plot(history.epoch, history.history['val_accuracy'], label = 'val_acc')
    axes[0].set_title('Accuracy vs Epochs', fontsize = 20)
    axes[0].set_xlabel('Epochs', fontsize = 15)
    axes[0].set_ylabel('Accuracy', fontsize = 15)
    axes[0].legend()

    axes[1].plot(history.epoch, history.history['loss'], label = 'loss')
    axes[1].plot(history.epoch, history.history['val_loss'], label="val_loss")
    axes[1].set_title("Loss Curve",fontsize=18)
    axes[1].set_xlabel("Epochs",fontsize=15)
    axes[1].set_ylabel("Loss",fontsize=15)
    axes[1].legend()

    plt.show()
    
model_performance_graphs2()
```
